Every year, American high school students take SATs, which are standardized tests intended to measure literacy, numeracy, and writing skills. There are three sections - reading, math, and writing, each with a **maximum score of 800 points**. These tests are extremely important for students and colleges, as they play a pivotal role in the admissions process.

Analyzing the performance of schools is important for a variety of stakeholders, including policy and education professionals, researchers, government, and even parents considering which school their children should attend. 

You have been provided with a dataset called `schools.csv`, which is previewed below.

You have been tasked with answering three key questions about New York City (NYC) public school SAT performance.

```python
# Loading the necessary modules
import pandas as pd

# Read in the data
schools = pd.read_csv("schools.csv")

# Preview the data
schools.head()
```

```python
# Scoring the maximum possible score in math 
max_score = 800 
max_score
```

```python
# Range of best results
best = 0.8 * max_score
best
```

```python
# Checking all average per school

score_per_school = schools[["school_name", "average_math"]]
score_per_school
```

```python
# Display in descending ordered

score_per_school_sorted = score_per_school.sort_values(by="average_math", ascending=False)
score_per_school_sorted
```

```python
# Display all the best math results

best_math_schools = score_per_school_sorted[score_per_school_sorted["average_math"] >= best]
best_math_schools
```

```python
# Total SAT column

schools["total_SAT"] = schools["average_math"]+schools["average_reading"]+schools["average_writing"]
schools
```

```python
# Method B of summarizing three columns

schools['total_SAT'] = schools.apply(lambda row: row['average_math'] + row['average_reading'] + row['average_writing'], axis=1)
schools
```

```python
# Sorted dataframe based on combined SAT scores

schools_sorted = schools.sort_values(by="total_SAT", ascending=False)
schools_sorted
```

## Top 10 performing schools based on combined SAT scores - solution A

```python
top_10_schools = schools_sorted[["school_name", "total_SAT"]].head(10)
top_10_schools
```

## Top 10 performing schools based on combined SAT scores - solution B

```python
top_10_schools = schools.sort_values("total_SAT", ascending=False)[["school_name", "total_SAT"]].head(10)
top_10_schools
```

# Q2 : Which single borough has the largest standard deviation in the combined SAT score? - solution A

```python
# Calculate standard deviation per borough for total_SAT and create a new dataframe
std_dev_per_borough = schools.groupby("borough")["total_SAT"].std().reset_index(name="std_dev")
std_dev_per_borough
```

### Adding columns for the number of schools in the borough, the mean and the standard deviation of total_SAT

```python
# 1 - Number of schools

std_dev_per_borough["num_schools"] = schools.groupby("borough")["school_name"].count().reset_index(name="num_schools")["num_schools"]
std_dev_per_borough
```

```python
# Checking if the count of number of schools is correct

filter_borough = schools[schools["borough"]=="Staten Island"]
unique_schools_bronx = filter_borough["school_name"].unique()
unique_schools_bronx
```

```python
# 2 - Average_SAT 

std_dev_per_borough["average_SAT"] = schools.groupby("borough")["total_SAT"].mean().reset_index(name="average_SAT")["average_SAT"]
std_dev_per_borough
```

```python
# 3 - Standard deviation of total_SAT 

std_dev_per_borough["std_SAT"] = schools.groupby("borough")["total_SAT"].std().reset_index(name="std_SAT")["std_SAT"]
std_dev_per_borough
```

```python
# Round all values to two decimals

std_dev_per_borough[['std_dev', 'num_schools', 'average_SAT', 'std_SAT']] = std_dev_per_borough[['std_dev', 'num_schools', 'average_SAT', 'std_SAT']].round(2)
std_dev_per_borough
```

```python
# Ordering the dataframe obtained with highest std at the top

ordered_std_dev = std_dev_per_borough.sort_values(by="std_SAT", ascending = False)

# Set the colum "borough" as index
ordered_std_dev.set_index('borough', inplace=True)

# Selecting only the three columns 
largest_std_dev = ordered_std_dev[["num_schools", "average_SAT", "std_SAT"]][:1]

# Display the final results
largest_std_dev
```

# Q2 : Which single borough has the largest standard deviation in the combined SAT score? - solution B

```python
# Which NYC borough has the highest standard deviation for total_SAT?

boroughs = schools.groupby("borough")["total_SAT"].agg(["count", "mean", "std"]).round(2)
boroughs
```

```python
# Filter for max std and make borough a column

largest_std_dev = boroughs[boroughs["std"] == boroughs["std"].max()]
largest_std_dev
```

```python
# Rename the columns for clarity

largest_std_dev = largest_std_dev.rename(columns={"count": "num_schools", "mean": "average_SAT", "std": "std_SAT"})
largest_std_dev
```
